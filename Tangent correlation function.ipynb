```python
##to estimate tangent correlation and compare them with analytical results
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import optimize, special
from numpy import *
from pylab import *
import decimal
import matplotlib as mpl
import random
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
plt.style.use("seaborn-whitegrid")
b1=2
b2=0.5
b3=1
b4=4
Lp=50
n=150
ALL=1000
##load the data saved in D disc, give the values to lists named config1-4
config1=np.load('D:Configurations.npy')
config2=np.load('D:configurationsWithlb=0.5.npy')
config3=np.load('D:configurationsWithlb=1.npy')
config4=np.load('D:configurationsWithlb=4.npy')
##the analytical formula for tangent correlation functions, with different lb(result in different L)
x=np.arange(0,150,0.1)
Y1=(math.e)**((-b1/Lp)*x)
Y2=(math.e)**((-b2/Lp)*x)
Y3=(math.e)**((-b3/Lp)*x)
Y4=(math.e)**((-b4/Lp)*x)
##compute the tangent correlations by simulation
i=0
k=0
c1=0
c2=0
c3=0
c4=0
tt1=np.zeros(ALL+1,dtype='float64')
tt2=np.zeros(ALL+1,dtype='float64')
tt3=np.zeros(ALL+1,dtype='float64')
tt4=np.zeros(ALL+1,dtype='float64')
TC1=np.zeros(n+1,dtype='float64')
TC2=np.zeros(n+1,dtype='float64')
TC3=np.zeros(n+1,dtype='float64')
TC4=np.zeros(n+1,dtype='float64')

I=np.zeros(n+1,dtype='float64')
while i<149:

    while k<ALL:
        t1=config1[k][0]    
        T1=config1[k][i+1]
        tt1[k]=np.dot(t1,T1)
        c1=c1+tt1[k]
        
        t2=config2[k][0]    
        T2=config2[k][i+1]
        tt2[k]=np.dot(t2,T2)
        c2=c2+tt2[k]
        
        t3=config3[k][0]    
        T3=config3[k][i+1]
        tt3[k]=np.dot(t3,T3)
        c3=c3+tt3[k]
        
        t4=config4[k][0]    
        T4=config4[k][i+1]
        tt4[k]=np.dot(t4,T4)
        c4=c4+tt4[k]     
        
        k=k+1      
    else:
        TC1[i]=c1/1000
        TC2[i]=c2/1000
        TC3[i]=c3/1000
        TC4[i]=c4/1000
        
        k=0
        I[i]=i
        i=i+1
        c1=0
        c2=0
        c3=0
        c4=0
else:    
    plt.scatter(I,TC2,c='b',s=1,label="b=0.5")
    plt.scatter(I,TC3,c='green',s=1,label="b=1")
    plt.scatter(I,TC1,c='r',s=1,label="b=2")    
    plt.scatter(I,TC4,c='orange',s=1,label="b=4")
    
    
    
    plt.plot(x,Y1,c='k',linestyle="-",alpha=0.5,label="analytical results")
    plt.plot(x,Y2,c='k',linestyle="-",alpha=0.5)
    plt.plot(x,Y3,c='k',linestyle="-",alpha=0.5)
    plt.plot(x,Y4,c='k',linestyle="-",alpha=0.5)
    plt.xlabel('i')
    plt.ylabel('Tangent Correlation')
    legend(loc='upper right')
    print(I)
    print(TC1)
    print(TC2)
    print(TC3)
    print(TC4)
plt.show()
```
