```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import optimize, special
from numpy import *
from pylab import *
import decimal
import matplotlib as mpl
import random
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
plt.style.use("seaborn-whitegrid")
np.seterr(divide='ignore', invalid='ignore')
config=np.load('D:configurationswithdifferanetlb50.npy')

x=np.arange(0,825,0.1)
LP=50
lb=0.5


i=0
k=0
Q=0
n=150


c=np.zeros(3,dtype='float64')
r=np.array([[[0]*3]*1000]*50,dtype='float64')
samples=np.array([[0]*2]*50,dtype='float64')
L=np.zeros(50,dtype='float64')
R=np.zeros(50,dtype='float64')
while Q<50:
    
    while k<1000:
        while i<n: 
            b=lb*config[Q][k][i]
            c=c+b
            i=i+1
        else:
            i=0
            r[Q][k]=c
            c=np.array([0,0,0],dtype='float64')
            k=k+1        
    else:
        k=0
        Q=Q+1
        lb=lb+0.1
        
else:
    Q=0
k=0
M=0
m=0
lb=0.5
while Q<50:
    while k<1000:    
        T=(r[Q][k])
        t=np.dot(T,T)
        M=M+t
        m=(M/1000)
                
        k=k+1
    else:
        k=0
        
        w=150*lb      
        L[Q]=w      
        R[Q]=math.sqrt(m)/w
        samples[Q][0]=L[Q]
        samples[Q][1]=R[Q]
        lb=lb+0.1
        m=0
        M=0
        Q=Q+1
else:
    print(samples)
    plt.scatter(L,R,s=5,c='r')
    plt.plot(x,((2*LP*x-2*(LP**2)*(1-(math.e)**(-x/LP)))**(1/2))/x)
    plt.xlabel('L (nm)')
    plt.ylabel('relative extenton (nm**2)')
    plt.savefig("D:/relative extenton.png")
plt.show()
```
